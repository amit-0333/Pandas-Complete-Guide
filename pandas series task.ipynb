###`Q-2:` Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

````python
# code here
import pandas as pd

# Creating two Series
s1 = pd.Series([10, 20, 30, 40])
s2 = pd.Series([1, 2, 3, 4])

# Addition
print("Addition:")
print(s1 + s2)

# Subtraction
print("\nSubtraction:")
print(s1 - s2)

# Multiplication
print("\nMultiplication:")
print(s1 * s2)

# Division
print("\nDivision:")
print(s1 / s2)
````

````python
import pandas as pd


# TC: O(n)
# SC: O(n)

# APPROACH:
# 1. Create a function that takes two Series.
# 2. Perform arithmetic operations.
# 3. Print results one by one.


def series_operations(s1, s2):

    # Addition
    print("Addition:")
    print(s1 + s2)

    # Subtraction
    print("\nSubtraction:")
    print(s1 - s2)

    # Multiplication
    print("\nMultiplication:")
    print(s1 * s2)

    # Division
    print("\nDivision:")
    print(s1 / s2)


# Creating Series
s1 = pd.Series([10, 20, 30, 40])
s2 = pd.Series([1, 2, 3, 4])

# Function Call
series_operations(s1, s2)
````

###`Q-3` Write a Pandas program to compare the elements of the two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]



````python
# code here
import pandas as pd


# TC: O(n)
# SC: O(n)

# APPROACH:
# 1. Create two Pandas Series.
# 2. Compare elements using comparison operators.
# 3. Print comparison results.


def compare_series(s1, s2):

    # Equal comparison
    print("Equal:")
    print(s1 == s2)

    # Greater than comparison
    print("\nGreater Than:")
    print(s1 > s2)

    # Less than comparison
    print("\nLess Than:")
    print(s1 < s2)


# Creating Series
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([1, 3, 2, 4, 6])

# Function Call
compare_series(s1, s2)
````

###`Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
# code here
import pandas as pd


# TC: O(n)
# SC: O(n)

# APPROACH:
# 1. Create a function that takes a Series and target datatype.
# 2. Use astype() to convert datatype.
# 3. Return converted Series.


def change_dtype(series, dtype):

    # Convert datatype
    converted_series = series.astype(dtype)

    return converted_series


# Creating Series
s = pd.Series([1, 2, 3, 4])

# Function Call
result = change_dtype(s, "float")

print(result)
print("\nDatatype:", result.dtype)
````

````python
import pandas as pd

# Read data
item = pd.read_csv('/content/items.csv')

# Convert $ to float
item['item_price'] = (
    item['item_price']
    .str.replace('$', '', regex=False)
    .astype(float)
)

# Fill missing values with mean
item['item_price'] = item['item_price'].fillna(item['item_price'].mean())

# Final mean
print(item['item_price'].mean())
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
item = pd.read_csv('/content/items.csv')

# Clean price column
item['item_price'] = (
    item['item_price']
    .str.replace('$', '', regex=False)
    .astype(float)
)

# i) 30th and 6th percentile
p30 = item['item_price'].quantile(0.30)
p6 = item['item_price'].quantile(0.06)

print("30th percentile:", p30)
print("6th percentile:", p6)

# ii) Histogram (bin size = 50)
plt.hist(item['item_price'], bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# iii) Count items between 1000 and 2000
count_range = item[(item['item_price'] >= 1000) & (item['item_price'] <= 2000)].shape[0]

print("Items between 1000 and 2000:", count_range)
````
